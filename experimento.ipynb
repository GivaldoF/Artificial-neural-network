```
import pandas as pd
```

Importando os dados


```
pacientes_radiomics = pd.read_csv("pacientes_radiomics.csv")
```

Removendo linhas sem valores

```
pacientes_radiomics = pacientes_radiomics.dropna(how="any").reset_index(drop=True)
pacientes_radiomics

```

Criando o df das classes

```
targets = pacientes_radiomics.classe.copy()
targets.replace({'M' : 1, 'B': 0}, inplace=True)
```

Selecionando os atributos dos tipos: shape, glcm, glrlm, glszm

```
new_pacientes_radiomics = pacientes_radiomics.filter(regex="shape|glcm|glrlm|glszm", axis=1)
```

Normalizando os atributos usando MinMax

```
from sklearn.preprocessing import MinMaxScaler
dataframe_normalizado = pd.DataFrame()

scaler = MinMaxScaler()
for col in new_pacientes_radiomics.columns: 
  coluna_array_normalizada = scaler.fit_transform(new_pacientes_radiomics[[col]].to_numpy())
  coluna_normalizada = pd.DataFrame(coluna_array_normalizada)
  dataframe_normalizado = pd.concat([dataframe_normalizado, coluna_normalizada], axis=1)

# recuperando os nomes das colunas

dataframe_normalizado.columns = new_pacientes_radiomics.columns

dataframe_normalizado.head()
```

Começar a treinar os modelos

```
from sklearn.model_selection import train_test_split

from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier

from sklearn.metrics import roc_auc_score
from sklearn.metrics import accuracy_score

from sklearn.model_selection import cross_val_predict
from sklearn.model_selection import LeaveOneOut
```

Divivindo os dados

```
X = dataframe_normalizado
y = targets
#X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=2/3, random_state=42, stratify=targets)

loo = LeaveOneOut()
```

KNN 

```
knn = KNeighborsClassifier()
y_pred = cross_val_predict(knn, X, y, cv=loo,n_jobs=-1)
auc = roc_auc_score(y, y_pred)
accuracy = accuracy_score(y, y_pred)
print("Métricas: ")
print("Auc: {}".format(auc))
print("Accuracy: {}".format(accuracy))
```

Naive Bayes

```
gnb = GaussianNB()
y_pred = cross_val_predict(gnb, X, y, cv=loo,n_jobs=-1)
auc = roc_auc_score(y, y_pred)
accuracy = accuracy_score(y, y_pred)
print("Métricas: ")
print("Auc: {}".format(auc))
print("Accuracy: {}".format(accuracy))
```

SVM

```
svc = SVC()
y_pred = cross_val_predict(svc, X, y, cv=loo,n_jobs=-1)
auc = roc_auc_score(y, y_pred)
accuracy = accuracy_score(y, y_pred)
print("Métricas: ")
print("Auc: {}".format(auc))
print("Accuracy: {}".format(accuracy))
```

C5.0

```
c50 = DecisionTreeClassifier()
y_pred = cross_val_predict(c50, X, y, cv=loo,n_jobs=-1)
auc = roc_auc_score(y, y_pred)
accuracy = accuracy_score(y, y_pred)
print("Métricas: ")
print("Auc: {}".format(auc))
print("Accuracy: {}".format(accuracy))
```

Random Forest


```
rfc = RandomForestClassifier()
y_pred = cross_val_predict(rfc, X, y, cv=loo,n_jobs=-1)
auc = roc_auc_score(y, y_pred)
accuracy = accuracy_score(y, y_pred)
print("Métricas: ")
print("Auc: {}".format(auc))
print("Accuracy: {}".format(accuracy))
```

MLP

```
mlpc = MLPClassifier()
y_pred = cross_val_predict(mlpc, X, y, cv=loo,n_jobs=-1)
auc = roc_auc_score(y, y_pred)
accuracy = accuracy_score(y, y_pred)
print("Métricas: ")
print("Auc: {}".format(auc))
print("Accuracy: {}".format(accuracy))
```
